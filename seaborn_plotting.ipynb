```python
import pandas as pd
df = pd.read_csv('menu.csv')
df
```

```python
# general info about the data set
df.describe()
```

```python
# which food has maximum of sodium
most_salty = df.loc[df['Sodium'] == df['Sodium'].max()]['Item']
print(most_salty)

# or use df.idxmax()
most_salty1 = df['Sodium'].idxmax()
print(most_salty1)
```

```python
# plot the sodium amount in each category
# sns.swarmplot
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(15, 6))
plot = sns.swarmplot(x='Category', y='Sodium', data=df, s=3, palette='deep')
plot.set_xticklabels(plot.get_xticklabels(), rotation=45, horizontalalignment='right')
plt.show()
```

```python
#sns.jointplot kind=default
joint_default = sns.jointplot(x='Protein', y='Total Fat', height=8, data=df )
plt.show()
```

```python
# sns.jointplot kind='hex'
joint_hex = sns.jointplot(x='Protein', y='Total Fat', data=df, height=8, kind='hex')
plt.show()
```

```python
# sns.jointplot kind='reg' 
joint_reg = sns.jointplot(x='Protein', y='Total Fat', height=8, ratio=3, data=df, kind='reg') # height ratio setting for 
plt.show()
```

```python
# use JointGrid parameter to control the size and layout of the figure
joint = sns.JointGrid(data=df, x='Protein', y='Total Fat', palette='flare',hue='Category', height=8, ratio=3, marginal_ticks=False)
joint.plot(sns.scatterplot, sns.histplot)
plt.show()
```

```python
# use JointGrid parameter to control the size and layout of the figure, ## marginal_tcks set to True
joint = sns.JointGrid(data=df, x='Protein', y='Total Fat', palette="crest", hue='Category', height=8, ratio=3, marginal_ticks=True)
joint.plot(sns.scatterplot, sns.histplot)
plt.show()
```

```python
# with jointgrid, plotting becomes more flexible to plot ax_marg_x and ax_marg_y with different plotting style
joint = sns.JointGrid(marginal_ticks=True, height=8)
x, y = df['Protein'], df['Calories']
sns.scatterplot(x=x, y=y, s=70, linewidth=1.5, ax=joint.ax_joint)
sns.histplot(x=x, fill=False, linewidth=2, ax=joint.ax_marg_x)
sns.kdeplot(y=y, linewidth=2, ax=joint.ax_marg_y)
plt.show()
```

```python
# boxplot
plt.figure(figsize=(12, 8))
sns.set_style('whitegrid')
sns.boxplot(x='Category', y='Dietary Fiber', palette='colorblind', saturation=0.7, data=df)

plt.show()
```
